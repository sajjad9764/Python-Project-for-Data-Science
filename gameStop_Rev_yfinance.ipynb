```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for GameStop
ticker_symbol = "GME"

# Initialize the Ticker object
gamestop = yf.Ticker(ticker_symbol)

# Fetch annual income statements
annual_income_stmt = gamestop.financials
print("Annual Income Statement for GameStop:")
print(annual_income_stmt)

# Extract revenue from the annual income statement
annual_revenue = annual_income_stmt.loc['Total Revenue']
print("\nAnnual Revenue for GameStop:")
print(annual_revenue)

# Fetch quarterly income statements
quarterly_income_stmt = gamestop.quarterly_financials
print("\nQuarterly Income Statement for GameStop:")
print(quarterly_income_stmt)

# Extract revenue from the quarterly income statement
quarterly_revenue = quarterly_income_stmt.loc['Total Revenue']
print("\nQuarterly Revenue for GameStop:")
quarterly_revenue.head()
```
