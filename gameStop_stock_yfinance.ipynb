```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for GameStop
ticker_symbol = "GME"

# Initialize the Ticker object
gamestop = yf.Ticker(ticker_symbol)

# Fetch historical market data for the past year
historical_data = gamestop.history(period="1y")
print("Historical Stock Prices for GameStop:")
print(historical_data)

# Fetch annual income statements
income_stmt = gamestop.financials
print("\nAnnual Income Statement:")
print(income_stmt)

# Fetch quarterly income statements
quarterly_income_stmt = gamestop.quarterly_financials
print("\nQuarterly Income Statement:")
print(quarterly_income_stmt)

# Fetch the balance sheet
balance_sheet = gamestop.balance_sheet
print("\nBalance Sheet:")
print(balance_sheet)

# Fetch the cash flow statement
cash_flow = gamestop.cashflow
print("\nCash Flow Statement:")
print(cash_flow)

# Fetch key metrics
key_metrics = gamestop.key_metrics
print("\nKey Metrics:")
key_metrics.head()
```
